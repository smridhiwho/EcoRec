```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Simulated product attributes dataset (replace with your actual dataset)
df3=df[['id','name','categories','reviews.text']]
df3 = df3.dropna(subset=['reviews.text'])
df3=df3.dropna(subset=['name'])
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df3['reviews.text'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

#Step 6: Recommendation with Hybrid Approach:


```python
def get_content_based_recommendations(product_id):
    try:
        # Find the index of the queried product in the product DataFrame
        product_idx = df3[df3['name'] == product_id].index[0]

        # Calculate content-based recommendations based on cosine similarity
        cosine_sim_scores = list(enumerate(cosine_sim[product_idx]))
        cosine_sim_scores = sorted(cosine_sim_scores, key=lambda x: x[1], reverse=True)

        # Get top similar products (excluding the queried product)
        content_based_recommendations = [df3.iloc[i[0]]['name'] for i in cosine_sim_scores if i[0] != product_idx]

        return content_based_recommendations

    except IndexError:
        print(f"Product ID '{product_id}' not found in the product_data DataFrame.")
        return []

```

```python
def hybrid_recommendation(user_id, product_id):
    # Collaborative Filtering
    predicted_rating = algo.predict(user_id, product_id).est

    # Content-Based Filtering
    product_idx = df3[df3['name'] == product_id].index
    content_based_score = cosine_sim[product_idx]  # Calculate content-based score

    # Sentiment Analysis
    sentiment_score = df4[df4['name'] == product_id]['sentiment'].mean()

    # Real-Time Trends
    is_trending = product_id in trending_products

    # Combine Scores
    hybrid_score = 0.4 * predicted_rating + 0.2 * content_based_score + \
                   0.2 * sentiment_score + 0.2 * is_trending

    # Get content-based recommendations
    content_based_recommendations = get_content_based_recommendations(product_id)

    return hybrid_score, content_based_recommendations

user_id = 'AVqkIhwDv8e3D1O-lebb	'
product_id = 'Amazon Kindle Lighted Leather Cover,,,\nKindle Keyboard,,,'
hybrid_score, content_based_recos = hybrid_recommendation(user_id, product_id)
print("Hybrid Recommendation Score:", hybrid_score)
print("Top 5 Unique Content-Based Recommendations:")
unique_recos = set()
idx = 1

for reco in content_based_recos:
    if reco not in unique_recos:
        print(f"{idx}. {reco}")
        unique_recos.add(reco)
        idx += 1

        if idx > 5:  # Stop after printing 5 unique recommendations
            break
```

The hybrid score and content-based recommendations are both parts of the output of the hybrid_recommendation function. They provide information about the recommendation for a specific user and product.

## Hybrid Score:

The hybrid score is a numerical value that represents the overall recommendation score for a specific user and product combination.
It's calculated by combining multiple factors, such as collaborative filtering predictions, content-based similarity scores, sentiment analysis scores, real-time trend presence, and potentially other factors.
A higher hybrid score indicates a stronger recommendation for the user and product.
## Content-Based Recommendations:

The content_based_recos variable contains a list of recommended product IDs based on content-based filtering.
These recommendations are generated by finding products that are similar to the queried product based on their attributes, descriptions, or other relevant features.
These recommendations are not personalized to the user but are purely based on the attributes and similarities between products.
Together, the hybrid score and content-based recommendations provide insights into the suitability of a product for a specific user. The hybrid score incorporates multiple factors, including the user's interactions, sentiment analysis, and real-time trends, while the content-based recommendations focus solely on similarities between products' attributes.

When interpreting the output, you can consider the hybrid score to prioritize products that are not only similar to what the user has shown interest in but also align with sentiment and trend considerations. Content-based recommendations can serve as alternatives that have similar attributes to the queried product, which might be useful for diversifying recommendations or offering choices that are different from the user's historical preferences.

Remember that the actual interpretation and usage of these scores and recommendations might vary based on the specific goals of your recommendation system and the preferences of your users.

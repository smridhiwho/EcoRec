```python
#list for eco_friendly
ecof=[False, True, False, False, False, False, True, True, True, True, False, False, True, True, False, True, False, True, False, False, False, False, True, False, True, False, True, False, False, True, True, False, True, False, False, True, False, True, True, False, False, False, True, True, True, True, False, False]
#list for socila responsibility
srep=[False, False, True, True, True, True, False, True, False, True, True, True, True, True, True, True, True, False, True, False, True, False, True, False, False, False, False, True, True, True, True, False, True, False, True, True, True, False, True, False, True, False, False, True, True, False, True, True]
```

```python
#eco friendly dictionary
ecof_dict = {}
for key in lis:
    for value in ecof:
        ecof_dict[key] = value
        ecof.remove(value)
        break

# Printing resultant dictionary
print("Resultant dictionary is : " + str(ecof_dict))

#social responsibility dictionary
srep_dict = {}
for key in lis:
    for value in srep:
        srep_dict[key] = value
        srep.remove(value)
        break

# Printing resultant dictionary
print("Resultant dictionary is : " + str(srep_dict))
```

```python
df['eco_friendly'] = df['name'].map(ecof_dict)
df['social_responsibility'] = df['name'].map(srep_dict)
df
```

```python
def calculate_sustainability_score(eco_friendly, social_responsibility):
    eco_weight = 0.6
    social_weight = 0.4
    # Calculate sustainability score
    score = eco_weight * eco_friendly + social_weight * social_responsibility
    return score

# Define weight for the sustainability score itself

# Calculate and assign weighted sustainability scores
df['eco_score'] = df.apply(
    lambda row: calculate_sustainability_score(row['eco_friendly'], row['social_responsibility']),
    axis=1
)
```
